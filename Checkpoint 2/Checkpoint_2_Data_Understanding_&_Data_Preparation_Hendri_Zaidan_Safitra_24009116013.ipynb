**MENCARI OUTLIERS PADA DATASET**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

```python
df
```

Diperlihatkan presentase yang tidak terlalu tinggi tetapi saya tetap imputasi karena "Miles" merupakan variabel penting untuk analisis. Metode imputasinya yaitu dengan menggunakan Winsorizing yang merupakan mempertahankan semua data tanpa membuang nilai ekstrem.

```python
columns_to_impute = ["MILES"]

for col in columns_to_impute:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Menggunakan .loc[] agar tidak muncul SettingWithCopyWarning
    df.loc[:, col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

Melakukan pengecekan ulang pada dataset setelah diimputasi

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

**CONSTRUCT DATA**

Disini saya menambahkan kolom yaitu waktu yang ditempuh dan kategori jaraknya karena untuk memudahkan menganalisis sebuah dataset

```python
df["TRIP_DURATION"] = (pd.to_datetime(df["END_DATE"]) - pd.to_datetime(df["START_DATE"])).dt.total_seconds() / 60


df["DISTANCE_CATEGORY"] = df["MILES"].apply(lambda x: "Short" if x < 5 else "Medium" if x <= 20 else "Long")

print(df[["MILES", "TRIP_DURATION", "DISTANCE_CATEGORY"]].head())
```

Mengecek kembali untuk outlinernya pada kolom TRIP_DURATION

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Diperlihatkan presentase yang tidak terlalu tinggi tetapi saya tetap imputasi karena "TRIP_DURATION" merupakan variabel penting untuk analisis. Metode imputasinya yaitu dengan menggunakan Winsorizing yang merupakan mempertahankan semua data tanpa membuang nilai ekstrem

```python
columns_to_impute = ["TRIP_DURATION"]

for col in columns_to_impute:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Winsorizing: Memotong outlier tanpa menghapus data
    df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

Melakukan pengecekan ulang pada dataset setelah diimputasi

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```
